```python
import pandas as pd
```

```python
#Datasetin csv dosyası olarak okunması
veri=pd.read_csv('student-mat.csv', sep=';')
```

```python
#İstenilen özelliklerin seçilmesi.
veri=veri[['G1','G2','G3','studytime','failures','absences','age']]
```

```python
veri.head(11)
```

```python
#Sutun adlarının türkçelştirilmesi ve satırlarda ki verierin korunması
veri.rename(columns={'G1': 'Vize1','G2': 'Vize2','G3': 'Final',
                     'studytime': 'Calisma_Suresi','failures': 'Sinif_Tekrari','absences': 'Devamsizlik','age': 'Yas'},
           inplace=True)
```

```python
veri.head(11)
```

```python
veri.dtypes
```

```python
#Girdi ve çıktı değerlerinin oluşturulması.
import numpy as np
y=np.array(veri['Final'])
X=np.array(veri.drop('Final',axis=1))
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Eğitim ve test verilerinin bölünmesi.
X_egitim, X_test, y_egitim, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
#Doğrusal Regresyon
from sklearn.linear_model import LinearRegression
```

```python
linear=LinearRegression()
```

```python
linear.fit(X_egitim,y_egitim)
```

```python
#Model ½83 oranında doğru tahmin yapıyor
linear.score(X_test,y_test)
```

```python
linear.score(X_egitim,y_egitim)
```

```python
#Test oranı ve egitim oranının birbirine yakın olması beklenir
```

```python
print('Katsayılar: \n',linear.coef_)
print('Sabit: \n', linear.intercept_)
```
